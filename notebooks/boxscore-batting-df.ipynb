```python
import json
import pandas as pd

# Load the boxscore JSON
with open('assets/json/boxscores/mlb_boxscores_2025-07-05.json', 'r') as f:
    boxscore = json.load(f)

batting_rows = []

for game in boxscore:
    game_boxscore = game['boxscore']
    for team_type in ['home', 'away']:
        team = game_boxscore['teams'][team_type]
        batters = team.get('batters', [])
        players = team.get('players', {})

        for batter_id in batters:
            player_key = f"ID{batter_id}"
            player = players.get(player_key)
            if not player:
                continue
            person = player['person']
            stats = player.get('stats', {}).get('batting', {})
            row = {
                'game_id': game['gamePk'],
                'team': team['team']['id'],
                'player_id': person['id'],
                'batting_order': player.get('battingOrder', None),
                'summary': stats.get('summary', None),
                'gamesPlayed': stats.get('gamesPlayed', None),
                'flyouts': stats.get('flyOuts', None),
                'groundOuts': stats.get('groundOuts', None),
                'airouts': stats.get('airOuts', None),
                'runs': stats.get('runs', None),
                'doubles': stats.get('doubles', None),
                'triples': stats.get('triples', None),
                'homeRuns': stats.get('homeRuns', None),
                'strikeOuts': stats.get('strikeOuts', None),
                'baseOnBalls': stats.get('baseOnBalls', None),
                'intentionalWalks': stats.get('intentionalWalks', None),
                'hits': stats.get('hits', None),
                'hitByPitch': stats.get('hitByPitch', None),
                'atBats': stats.get('atBats', None),
                'caughtStealing': stats.get('caughtStealing', None),
                'stolenBases': stats.get('stolenBases', None),
                'groungdIntoDoublePlay': stats.get('groundedIntoDoublePlay', None),
                'groundedIntoTriplePlay': stats.get('groundedIntoTriplePlay', None),
                'plateAppearances': stats.get('plateAppearances', None),
                'leftOnBase': stats.get('leftOnBase', None),
                'totalBases': stats.get('totalBases', None),
                'rbi': stats.get('rbi', None),            
                'leftOnBase': stats.get('leftOnBase', None)
            }
            batting_rows.append(row)

df = pd.DataFrame(batting_rows)

# Optionally, save all boxscores to a CSV file
df.to_csv('assets/csv/boxscores/mlb_boxscores_2025-07-05.csv', index=False)
```
